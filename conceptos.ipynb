
# Conceptos de programacion en Redes, enfocado a chats


### ¿Qué es un socket en redes? ¿Qué tipos hay y cuál estás usando acá? ¿Qué significa AF_INET y SOCK_STREAM?
- Un socket es una forma de comunicación entre dos clientes o entidades, de forma que se puedan enviar y recibir datos de forma ordenada actualizando los datos de cada uno de los mensajes enviados por medio del servidor que envia las actualizaciones de cada cliente a todos los clientes
